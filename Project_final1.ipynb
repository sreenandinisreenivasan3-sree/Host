```python
# Define features (X) and target (y)
X = df.drop('placed', axis=1)
y = df['placed']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
num_cols = X_train.select_dtypes(include=["int64","float64","bool"]).columns
cat_cols = X_train.select_dtypes(include=["object","category"]).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ])
```

```python
# Decision Tree pipeline
dt_pipe = ImbPipeline([
    ("prep", preprocess),
    ("select", SelectKBest(score_func=f_classif, k=20)),
    ("smote", SMOTE(random_state=42)),
    ("model", DecisionTreeClassifier(
        random_state=42, max_depth=6, min_samples_split=10, min_samples_leaf=5
    ))
])
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

# Decision Tree
dt_pipe.fit(X_train, y_train)
dt_pred = dt_pipe.predict(X_test)
dt_proba = dt_pipe.predict_proba(X_test)[:, 1]
print("=== Decision Tree ===")
print(confusion_matrix(y_test, dt_pred))
print(classification_report(y_test, dt_pred))
print("Accuracy:", accuracy_score(y_test, dt_pred))
print("ROC AUC:", roc_auc_score(y_test, dt_proba))
```

```python
#  Random Forest pipeline
rf_pipe = ImbPipeline([
    ("prep", preprocess),
    ("select", SelectKBest(score_func=f_classif, k=20)),
    ("smote", SMOTE(random_state=42)),
    ("model", RandomForestClassifier(
        n_estimators=300, random_state=42, n_jobs=-1
    ))
])
```

```python
# Random Forest
rf_pipe.fit(X_train, y_train)
rf_pred = rf_pipe.predict(X_test)
rf_proba = rf_pipe.predict_proba(X_test)[:, 1]
print("\n=== Random Forest ===")
print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
print("Accuracy:", accuracy_score(y_test, rf_pred))
print("ROC AUC:", roc_auc_score(y_test, rf_proba))
```

```python
#  XGBoost pipeline
xgb_pipe = ImbPipeline([
    ("prep", preprocess),
    ("select", SelectKBest(score_func=f_classif, k=20)),
    ("smote", SMOTE(random_state=42)),
    ("model", XGBClassifier(
        objective="binary:logistic", eval_metric="auc",
        random_state=42, n_jobs=-1
    ))
])
```

```python
# XGBoost
xgb_pipe.fit(X_train, y_train)
xgb_pred = xgb_pipe.predict(X_test)
xgb_proba = xgb_pipe.predict_proba(X_test)[:, 1]
print("\n=== XGBoost ===")
print(confusion_matrix(y_test, xgb_pred))
print(classification_report(y_test, xgb_pred))
print("Accuracy:", accuracy_score(y_test, xgb_pred))
print("ROC AUC:", roc_auc_score(y_test, xgb_proba))
```

```python
import json
from typing import Dict, Union

class PlacementPredictor:
    def __init__(self, model_path='placement_model.pkl', 
                 artifacts_path='model_artifacts.json',
                 dtypes_path='column_dtypes.json'):
        """
        Initialize the predictor with trained model and artifacts
        """
        self.model = joblib.load(model_path)
        
        with open(artifacts_path, 'r') as f:
            self.artifacts = json.load(f)
        
        with open(dtypes_path, 'r') as f:
            self.column_dtypes = json.load(f)
    
    def validate_input(self, input_data: Dict) -> bool:
        """
        Validate input data has all required features with correct types
        """
        # Check all required features are present
        required_features = self.artifacts['feature_names']
        missing_features = [f for f in required_features if f not in input_data]
        
        if missing_features:
            raise ValueError(f"Missing features: {missing_features}")
        
        # Basic type validation (customize based on your needs)
        for feature, value in input_data.items():
            if feature in self.artifacts['categorical_columns']:
                if not isinstance(value, str):
                    raise ValueError(f"Feature {feature} should be string/categorical")
            elif feature in self.artifacts['numerical_columns']:
                if not isinstance(value, (int, float)):
                    try:
                        float(value)
                    except:
                        raise ValueError(f"Feature {feature} should be numeric")
        
        return True
    
    def preprocess_input(self, input_data: Dict) -> pd.DataFrame:
        """
        Convert input dictionary to DataFrame with correct format
        """
        # Create DataFrame with a single row
        df = pd.DataFrame([input_data])
        
        # Ensure all columns are present (fill missing with None)
        for col in self.artifacts['feature_names']:
            if col not in df.columns:
                df[col] = None
        
        # Reorder columns to match training data
        df = df[self.artifacts['feature_names']]
        
        # Convert dtypes to match training data
        for col, dtype_str in self.column_dtypes.items():
            if col in df.columns:
                if 'float' in dtype_str:
                    df[col] = pd.to_numeric(df[col], errors='coerce')
                elif 'int' in dtype_str:
                    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
        
        return df
    
    def predict(self, input_data: Union[Dict, pd.DataFrame]) -> Dict:
        """
        Make prediction for a single input
        """
        # Validate and preprocess input
        if isinstance(input_data, dict):
            self.validate_input(input_data)
            processed_data = self.preprocess_input(input_data)
        else:
            processed_data = input_data
        
        # Make prediction
        prediction = self.model.predict(processed_data)[0]
        probability = self.model.predict_proba(processed_data)[0]
        
        # Prepare response
        response = {
            'prediction': int(prediction),
            'prediction_label': 'Placed' if prediction == 1 else 'Not Placed',
            'probability': {
                'not_placed': float(probability[0]),
                'placed': float(probability[1])
            },
            'confidence': float(max(probability)),
            'model_info': {
                'type': self.artifacts['model_type'],
                'accuracy': self.artifacts['performance_metrics']['accuracy'],
                'roc_auc': self.artifacts['performance_metrics']['roc_auc']
            }
        }
        
        return response
    
    def predict_batch(self, df: pd.DataFrame) -> pd.DataFrame:
        """
        Make predictions for multiple inputs
        """
        predictions = self.model.predict(df)
        probabilities = self.model.predict_proba(df)
        
        results = df.copy()
        results['prediction'] = predictions
        results['prediction_label'] = results['prediction'].map({1: 'Placed', 0: 'Not Placed'})
        results['probability_placed'] = probabilities[:, 1]
        results['probability_not_placed'] = probabilities[:, 0]
        
        return results

```
